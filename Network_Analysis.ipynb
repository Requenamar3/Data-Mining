```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/fenago/datasets/main/firenze.csv')
data.head()
```

```python
print(data.head())
print(data.info())
print(data.dtypes)
```

```python
import networkx as nx

# Convert the adjacency matrix into a list of edges
edges = []
for i in data.columns:
    for j in data.index:
        if data.loc[j, i] == 1:
            edges.append((j, i))

# Create the network graph from the list of edges
G = nx.Graph()
G.add_edges_from(edges)

# Remove self-loops if any
G.remove_edges_from(nx.selfloop_edges(G))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray')
plt.title("Network Graph of Florentine Families")
plt.show()
```

```python
# Degree Centrality
centrality = nx.degree_centrality(G)
print(centrality)
```
